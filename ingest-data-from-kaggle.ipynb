```python
from kaggle.api.kaggle_api_extended import KaggleApi

api = KaggleApi()
api.authenticate()  
```

```python
api.dataset_download_files(
    'augustop/portuguese-tweets-for-sentiment-analysis', 
    path='data/portuguese-tweets-for-sentiment-analysis', 
    unzip=True)

```

```python
import glob
import pandas as pd
paths = glob.glob('data/portuguese-tweets-for-sentiment-analysis/*.csv')
dfs = []
for path in paths:
    df = pd.read_csv(path)
    dfs.append(df)
    

dataset = pd.concat(dfs, ignore_index=True)
dataset
```

```python
dataset['sentiment'].replace({
    "Positivo": 1,
    "Negativo": -1,
    "Neutro": 0
}).unique()
```

```python
dataset = dataset.drop_duplicates(subset=['id'])
dataset = dataset.drop(columns=['tweet_date', 'id', 'query_used'])
dataset = dataset.rename(columns={"tweet_text": 'text'})
dataset = dataset.dropna()
dataset['sentiment'] = dataset['sentiment'].replace({
    "Positivo": 1,
    "Negativo": -1,
    "Neutro": 0
})

dataset
```

```python
dataset_dict = dataset.to_dict( orient='records')
dataset_dict
```
